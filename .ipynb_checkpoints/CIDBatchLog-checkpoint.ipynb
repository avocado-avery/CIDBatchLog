```python
import matplotlib.pyplot as plt
import numpy as np

# -----------------------------
# Measured HTLC values (from your tests)
# -----------------------------
HTLC_ETH_GAS_PER_CID = 229_567 / 1000        # kGas
HTLC_LOG_GAS_PER_CID = 194_319 / 1000        # kGas

# -----------------------------
# Batch measurements (REPLACE once measured)
# -----------------------------
batch_sizes = np.array([1, 5, 10, 25, 50, 100])

# Example placeholders — replace with REAL measurements
batch_total_gas = np.array([
    170_000,   # batch size 1
    230_000,   # batch size 5
    280_000,   # batch size 10
    360_000,   # batch size 25
    410_000,   # batch size 50
    447_463    # batch size 100 (your real value)
])

batch_gas_per_cid = (batch_total_gas / batch_sizes) / 1000  # kGas

# -----------------------------
# X range for HTLC lines
# -----------------------------
x_range = np.array([1, 100])

htlc_eth_line = np.full_like(x_range, HTLC_ETH_GAS_PER_CID, dtype=float)
htlc_log_line = np.full_like(x_range, HTLC_LOG_GAS_PER_CID, dtype=float)

# -----------------------------
# Plot
# -----------------------------
plt.figure(figsize=(12, 6))

plt.plot(
    x_range,
    htlc_eth_line,
    linestyle="--",
    linewidth=2,
    label="HTLC (ETH transfer)"
)

plt.plot(
    x_range,
    htlc_log_line,
    linestyle="--",
    linewidth=2,
    label="HTLC (logging-only)"
)

plt.plot(
    batch_sizes,
    batch_gas_per_cid,
    marker="o",
    linewidth=3,
    label="CIDBatchLog (batched)"
)

plt.xlabel("Number of CIDs")
plt.ylabel("Gas per CID (kGas)")
plt.title("Gas Efficiency vs Number of CIDs (Measured & Batched)")
plt.grid(True, linestyle="--", alpha=0.4)
plt.legend()
plt.tight_layout()
plt.show()
```
